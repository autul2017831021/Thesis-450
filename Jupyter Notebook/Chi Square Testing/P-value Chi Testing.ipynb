```python
scaled_all_song_correlated_features_csv_path = '../correlated_csv/scaled_all_song_correlated_features_csv.csv'
scaled_bangladesh_song_correlated_features_csv_path = '../correlated_csv/scaled_bangladesh_song_correlated_features_csv.csv'
scaled_kolkata_song_correlated_features_csv_path = '../correlated_csv/scaled_kolkata_song_correlated_features_csv.csv'
```

```python
alls = pd.read_csv(scaled_all_backup_csv_path)
bd = pd.read_csv(scaled_bangladesh_backup_csv_path)
ind = pd.read_csv(scaled_kolkata_backup_csv_path)
```

```python
corr_alls = pd.read_csv(scaled_all_song_correlated_features_csv_path)
corr_bd = pd.read_csv(scaled_bangladesh_song_correlated_features_csv_path)
corr_ind = pd.read_csv(scaled_kolkata_song_correlated_features_csv_path)
```

```python
def chi_testing(dataframe, index1, index2):
    fieldname1 = fieldnames[index1]
    fieldname2 = fieldnames[index2]
    dataset_table=pd.crosstab(dataframe[fieldname1],dataframe[fieldname2])
    observed = dataset_table.values
    chi2, pval, dof, expected = chi2_contingency(observed)
    return pval
```

```python
arr_of_dataframes = [alls,bd,ind]
arr_of_dataframes_name = ['All Songs','Bangladeshi Songs',"Indian Songs"]
for k in range(len(arr_of_dataframes)):
    print('\n -------------------------------------------------------------------------------------------------------------\n')
    for i in range(2,len(fieldnames)-1):
        for j in range(i+1,len(fieldnames)):
            p_val = chi_testing(arr_of_dataframes[k],i,j)
            if p_val <= 0.05:
                print('(',arr_of_dataframes_name[k],')',fieldnames[i],' X ',fieldnames[j],': p value is = ',p_val,'*There is a relation*'if p_val <= 0.05 else '','\n')
```

```python
def corr_chi_testing(dataframe, index1, index2):
    fieldname1 = corr_fieldnames[index1]
    fieldname2 = corr_fieldnames[index2]
    dataset_table=pd.crosstab(dataframe[fieldname1],dataframe[fieldname2])
    observed = dataset_table.values
    chi2, pval, dof, expected = chi2_contingency(observed)
    return pval
```

```python
corr_arr_of_dataframes = [corr_alls,corr_bd,corr_ind]
corr_arr_of_dataframes_name = ['All Songs','Bangladeshi Songs',"Indian Songs"]
for k in range(len(corr_arr_of_dataframes)):
    print('\n -------------------------------------------------------------------------------------------------------------\n')
    for i in range(2,3):
        for j in range(i+1,len(corr_fieldnames)):
            p_val = corr_chi_testing(corr_arr_of_dataframes[k],i,j)
            if p_val <= 0.05:
                print('(',corr_arr_of_dataframes_name[k],')',corr_fieldnames[i],' X ',corr_fieldnames[j],': p value is = ',p_val,'*There is a relation*'if p_val <= 0.05 else '','\n')
#             else :
#                 print('(',corr_arr_of_dataframes_name[k],')',corr_fieldnames[i],' X ',corr_fieldnames[j],': p value is = ',p_val,'*There is a relation*'if p_val <= 0.05 else '','\n')
```
